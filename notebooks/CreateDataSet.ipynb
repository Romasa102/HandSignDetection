# Create the directory for the datasets

This code would create the folder which would look like following.
```
NAME_OF_DATAFILE
|
|--ACTION1
|   |--1
|   |--2
|   |--3
|   |--NO_SEQUENCES
|
|--ACTION2
    |--1
    |--2
    |--3
    |--NO_SEQUENCES
```

````python
for action in config.ACTIONS:
    for sequence in range(config.NO_SEQUENCES):
        try:
            os.makedirs(os.path.join(DATA_PATH, action, str(sequence)))
        except:
            pass
````

# Data Collection


> [!WARNING]
Do not run this code when there is existing data inside the DATAFILE_NAME

````python
cap = cv2.VideoCapture(0)
with mp_holistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5) as model:
    for action in config.ACTIONS:
        for sequence in range(config.NO_SEQUENCES):
            for frame_num in range(config.SEQUENCE_LENGTH):
                ret, frame = cap.read()
                image, results = media_pipe_detection(frame, model)
                draw_landmarks(frame, results)

                if frame_num == 0:
                    cv2.putText(frame, 'STARTING COLLECTION', (120,200),
                                cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255, 0), 4, cv2.LINE_AA)
                    cv2.putText(frame, 'Collecting frames for {} Video Number {}'.format(action, sequence), (15,12),
                                cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,255), 1, cv2.LINE_AA)
                    cv2.imshow('Detection', frame)
                    cv2.waitKey(2000)
                else:
                    cv2.putText(frame, 'Collecting frames for {} Video Number {}'.format(action, sequence), (15,12),
                                cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,255), 1, cv2.LINE_AA)
                    cv2.imshow('Detection', frame)

                keypoints = extract_keypoints(results)
                npy_path = os.path.join(DATA_PATH, action, str(sequence), str(frame_num))
                np.save(npy_path, keypoints)

                if cv2.waitKey(10) & 0xFF == ord('q'):
                    break

    cap.release()
    cv2.destroyAllWindows()
````
